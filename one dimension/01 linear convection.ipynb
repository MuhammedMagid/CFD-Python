```python
#import needed libraries
import numpy  as np                     #here we load numpy
from matplotlib import pyplot      #here we load matplotlib
import time, sys                   #and load some utilities
#this makes matplotlib plots appear in the notebook (instead of a separate window)
%matplotlib inline

## Discretization
nx = 41
dx = 2 / (nx - 1)
nt = 25
dt = 0.025

# Initial condition (IC)
u = np.ones(nx)
for i in range(int(0.5 / dx), int(1 / dx + 1), 1):
    u[i] = 2
pyplot.plot(np.linspace(0, 2, nx), u,label='Initial Condition')

#solver (FDM)
un = np.ones(nx) #initialize a temporary array

for n in range(nt):  #loop for values of n from 0 to nt, so it will run nt times
    un = u.copy() ##copy the existing values of u into un
    for i in range(1, nx): ## you can try commenting this line and...
    #for i in range(nx): ## ... uncommenting this line and see what happens!
        u[i] = un[i] -  (dt / dx )* (un[i] - un[i-1])

pyplot.plot(np.linspace(0, 2, nx), u,label='After 25 time unit ');
pyplot.legend(fontsize='small')
```
